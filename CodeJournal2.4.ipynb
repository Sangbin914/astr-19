```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.triangular(-10,40,150,1000)

plt.figure(figsize=(10, 6))
plt.hist(x, bins=100)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Numbers From Triangular Distribution')
plt.grid()

plt.savefig("Code-Journal#2-4.pdf")
```
