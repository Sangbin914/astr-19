```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0,1,1000)

plt.figure(figsize=(10, 6))
plt.hist(x, bins=100)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers')
plt.grid()

plt.savefig("Code-Journal#2-3.pdf")
```
